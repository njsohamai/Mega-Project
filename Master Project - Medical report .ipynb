```python
import tkinter as tk
import customtkinter as ctk
import pandas as pd
from PIL import Image, ImageTk
import os
from datetime import datetime

def gen_report():
    global name,age,gender,fever,headache,joint_pain,bleeding,remarks,number,city,bp,sample,dengue,diabetes,vomit,hem
    name = name_entry.get()
    age = age_entry.get()
    gender = gender_entry.get()
    fever = fever_entry.get()
    headache ='Yes' if headache_entry.get().lower()=="yes" else 'No'
    joint_pain ='Yes' if joint_pain_entry.get().lower()=="yes" else 'No'
    bleeding ='Yes' if bleeding_entry.get().lower()=="yes" else 'No'
    remarks = remarks_entry.get()
    number = number_entry.get()
    city = city_entry.get()
    bp = bp_entry.get()
    sample = sample_entry.get()
    dengue = Dengue_entry.get()
    diabetes ='Yes' if diabetes_pain_entry.get().lower()=="yes" else 'No'
    vomit ='Yes' if vomit_entry.get().lower()=="yes" else 'No'
    hem = Hemoglobin_entry.get()

    filename = "Medical Data.xlsx"

    if not all([name,age,gender,fever,headache,joint_pain,bleeding,remarks,number,city,bp,sample,dengue,diabetes,vomit,hem]):
        tk.messagebox.showwarning('Input Error','Please fill in all the details!')
        return

    curr_dt = datetime.now().strftime('%d-%m-%y %H:%M:%S')
    if filename in os.listdir():
        df=pd.read_excel(filename)
        df.loc[len(df)]=[name,age,gender,fever,headache,joint_pain,bleeding,remarks,number,city,bp,sample,dengue,diabetes,vomit,hem,curr_dt]
        df.to_excel('Medical Data.xlsx',index=False)
    else:
        # Prepare data for Excel
        data = {
            "Name":[name],
            "Age":[age],
            "Gender":[gender],
            "Fever(C)":[fever],
            "Headache":[headache],
            "Joint Pain":[joint_pain],
            "Bleeding":[bleeding],
            "Doctor's Remark":[remarks],
            "Pateient Number":[number],
            "City":[city],
            "Blood Pressure":[bp],
            "Sample Type":[sample],
            "Dengue Fever Antibody":[dengue],
            "Diabetes":[diabetes],
            "Vomiting":[vomit],
            "Hemoglobin":[hem],
            "Date Time":[curr_dt]
        }
        df = pd.DataFrame(data)
        df.to_excel("Medical Data.xlsx",index=False)
    tk.messagebox.showinfo('Sucess','Medical Data is Sucessfully Entered.')

window = ctk.CTk()
window.title("Healthy Life Clinic-Medical Report Genderator")
window.geometry("600x600")

#Configure the appearance
ctk.set_appearance_mode("Dark")
ctk.set_default_color_theme("blue")

#Create title label(Hospital Name)
title_label=ctk.CTkLabel(window,text="Healthy Life Clinic",font=("Arial",45,"bold"))
title_label.place(relx=0.35,rely=0.05)

#Create 1 reow Label
name_label=ctk.CTkLabel(window,text="Patient Name",font=("Arial",15,"bold"))
name_label.place(relx=0.25,rely=0.25,anchor="e")
name_entry=ctk.CTkEntry(window)
name_entry.place(relx=0.255,rely=0.25,anchor="w")

age_label=ctk.CTkLabel(window,text="Patient Age",font=("Arial",15,"bold"))
age_label.place(relx=0.25,rely=0.35,anchor="e")
age_entry=ctk.CTkEntry(window)
age_entry.place(relx=0.255,rely=0.35,anchor="w")

gender_label=ctk.CTkLabel(window,text="Patient Gender",font=("Arial",15,"bold"))
gender_label.place(relx=0.25,rely=0.45,anchor="e")
gender_entry=ctk.CTkEntry(window)
gender_entry.place(relx=0.255,rely=0.45,anchor="w")

fever_label=ctk.CTkLabel(window,text="Fever (C)",font=("Arial",15,"bold"))
fever_label.place(relx=0.25,rely=0.55,anchor="e")
fever_entry=ctk.CTkEntry(window)
fever_entry.place(relx=0.255,rely=0.55,anchor="w")

headache_label=ctk.CTkLabel(window,text="Headache(Yes/No)",font=("Arial",15,"bold"))
headache_label.place(relx=0.25,rely=0.65,anchor="e")
headache_entry=ctk.CTkEntry(window)
headache_entry.place(relx=0.255,rely=0.65,anchor="w")

joint_pain_label=ctk.CTkLabel(window,text="Joint Pain(Yes/No)",font=("Arial",15,"bold"))
joint_pain_label.place(relx=0.25,rely=0.75,anchor="e")
joint_pain_entry=ctk.CTkEntry(window)
joint_pain_entry.place(relx=0.255,rely=0.75,anchor="w")

bleeding_label=ctk.CTkLabel(window,text="Bleeding(Yes/No)",font=("Arial",15,"bold"))
bleeding_label.place(relx=0.25,rely=0.85,anchor="e")
bleeding_entry=ctk.CTkEntry(window)
bleeding_entry.place(relx=0.255,rely=0.85,anchor="w")

remarks_label=ctk.CTkLabel(window,text="Doctor's Remarks",font=("Arial",15,"bold"))
remarks_label.place(relx=0.25,rely=0.95,anchor="e")
remarks_entry=ctk.CTkEntry(window)
remarks_entry.place(relx=0.255,rely=0.95,anchor="w")

#create 2 row labels
number_label=ctk.CTkLabel(window,text="Patient Number",font=("Arial",15,"bold"))
number_label.place(relx=0.55,rely=0.25,anchor="e")
number_entry=ctk.CTkEntry(window)
number_entry.place(relx=0.555,rely=0.25,anchor="w")

city_label=ctk.CTkLabel(window,text="Patient City",font=("Arial",15,"bold"))
city_label.place(relx=0.55,rely=0.35,anchor="e")
city_entry=ctk.CTkEntry(window)
city_entry.place(relx=0.555,rely=0.35,anchor="w")

bp_label=ctk.CTkLabel(window,text="Blood Pressure",font=("Arial",15,"bold"))
bp_label.place(relx=0.55,rely=0.45,anchor="e")
bp_entry=ctk.CTkEntry(window)
bp_entry.place(relx=0.555,rely=0.45,anchor="w")

sample_label=ctk.CTkLabel(window,text="Sample Type",font=("Arial",15,"bold"))
sample_label.place(relx=0.55,rely=0.55,anchor="e")
sample_entry=ctk.CTkEntry(window)
sample_entry.place(relx=0.555,rely=0.55,anchor="w")

Dengue_label=ctk.CTkLabel(window,text="Dengue Fever Anti-boody",font=("Arial",15,"bold"))
Dengue_label.place(relx=0.55,rely=0.65,anchor="e")
Dengue_entry=ctk.CTkEntry(window)
Dengue_entry.place(relx=0.555,rely=0.65,anchor="w")

diabetes_pain_label=ctk.CTkLabel(window,text="Diabetes(Yes/No)",font=("Arial",15,"bold"))
diabetes_pain_label.place(relx=0.55,rely=0.75,anchor="e")
diabetes_pain_entry=ctk.CTkEntry(window)
diabetes_pain_entry.place(relx=0.555,rely=0.75,anchor="w")

vomit_label=ctk.CTkLabel(window,text="Vomiting(Yes/No)",font=("Arial",15,"bold"))
vomit_label.place(relx=0.55,rely=0.85,anchor="e")
vomit_entry=ctk.CTkEntry(window)
vomit_entry.place(relx=0.555,rely=0.85,anchor="w")

Hemoglobin_label=ctk.CTkLabel(window,text="Hemoglobin",font=("Arial",15,"bold"))
Hemoglobin_label.place(relx=0.55,rely=0.95,anchor="e")
Hemoglobin_entry=ctk.CTkEntry(window)
Hemoglobin_entry.place(relx=0.555,rely=0.95,anchor="w")

submit_button=ctk.CTkButton(window,text="Generate Report",command=gen_report,width=180,height=160)
submit_button.place(relx=0.75,rely=0.45)

img_path=os.getcwd()+'\\logo.jpg'

pil_image=Image.open(img_path)
logo_img=ctk.CTkImage(light_image=pil_image,size=(100,100))
logo_label=ctk.CTkLabel(window,text='',image=logo_img)
logo_label.place(relx=0.25, rely=0.001)

window.state('zoomed')
window.resizable(True,False)
window.geometry("2024x2024px")
window.mainloop()
```

```python
from fpdf import FPDF

class MedicalReport(FPDF):
    def header(self):
        self.image(img_path,10,8,33)

        self.set_font('Arial','B',14)
        self.cell(0,10,'Healthy Life Hospital',border=False,ln=True,align='C')

        self.set_font('Arial','',10)
        self.cell(0,8,'123 Welness Street, Cityville, State, Zip-395582',border=False,ln=True,align='C')
        self.cell(0,8,'Phone:133 45678992 | Email: info@healthylifeclinic.com',border=False,ln=True,align='C')

        self.ln(10)
        self.set_draw_color(169,169,169)
        self.set_line_width(0.5)
        self.line(10,self.get_y(),200,self.get_y())
        self.ln(5)

    def footer(self):
        self.set_y(-15)
        self.set_font('Arial','I',8)
        self.cell(0,10,f'Page No. {self.page_no()} | This is a computer generated report. No signature is required')

    def patient_details(self,name,age,gender,number,city):
        self.set_font('Arial','B',10)
        self.set_fill_color(230,230,230)
        self.cell(0,10,'Personal Details',ln=True,align='L',fill=True)
        self.ln(2)

        self.set_font('Arial','',10)
        self.cell(40,10,'Name:',0,0)
        self.cell(0,10,name,0,1)
        self.cell(40,10,'Age:',0,0)
        self.cell(0,10,age,0,1)
        self.cell(40,10,'Gender:',0,0)
        self.cell(0,10,gender,0,1)
        self.cell(40,10,'Phone Number:',0,0)
        self.cell(0,10,number,0,1)
        self.cell(40,10,'City:',0,0)
        self.cell(0,10,city,0,1)
        self.ln(5)
    def clinical_observations(self,fever,headache,joint_pain,bleeding,bp,sample,dengue,diabetes,vomit,hem):
        self.set_font('Arial','B',10)
        self.set_fill_color(230,230,230)
        self.cell(0,10,'Clinical Observation',ln=True,align='L',fill=True)
        self.ln(2)

        self.set_font('Arial','',10)
        self.cell(40,10,'Fever:',0,0)
        self.cell(0,10,fever,0,1)
        self.cell(40,10,'Headache:',0,0)
        self.cell(0,10,headache,0,1)
        self.cell(40,10,'Joint Pain:',0,0)
        self.cell(0,10,joint_pain,0,1)
        self.cell(40,10,'Bleeding:',0,0)
        self.cell(0,10,bleeding,0,1)
        self.cell(40,10,'Blood Pressure:',0,0)
        self.cell(0,10,bp,0,1)
        self.cell(40,10,'Sample Type:',0,0)
        self.cell(0,10,sample,0,1)
        self.cell(40,10,'Dengue:',0,0)
        self.cell(0,10,dengue,0,1)
        self.cell(40,10,'Diabetes:',0,0)
        self.cell(0,10,diabetes,0,1)
        self.cell(40,10,'Vomit:',0,0)
        self.cell(0,10,vomit,0,1)
        self.cell(40,10,'Hemoglobin Level:',0,0)
        self.cell(0,10,hem,0,1)

    def remarks(self,remarks):
        self.set_font('Arial','B',10)
        self.set_fill_color(230,230,230)
        self.cell(0,10,'Doctors Remarks',ln=True,align='L',fill=True)
        self.ln(2)

        self.set_font('Arial','',10)
        self.multi_cell(0,10,remarks)
        self.ln(5)

pdf=MedicalReport()

pdf.add_page()
pdf.patient_details(name,age,gender,number,city)
pdf.clinical_observations(fever,headache,joint_pain,bleeding,bp,sample,dengue,diabetes,vomit,hem)
pdf.remarks(remarks)

pdf.output('Medical Report.pdf')
```
